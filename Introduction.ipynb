## Gelombang Seismik

Kalau kita sih sebagai manusia bebas memilih lintasannya, apakah mau yang paling dekat, atau paling cepat. Namun bagi gelombang seismik, dia harus tunduk (*sami'na wa ato'na*) pada hukum alam, yaitu hukum Snellius, yang merupakan representasi lain dari hukum Fermat (Azaz Fermat), yaitu azas lintasan tercepat. Jadi, gelombang seismik harus menjalar dari satu tempat ke tempat lain dengan "hanya" melalui lintasan yang paling cepat *thok til*. Ini saja pilihan baginya, *no choice!* Bersyukur kita sebagai geofisikawan, karena kalau gelombang seismik seperti manusia, wah bubar ilmu Geofisika ini.

Namun dalam metode seismik dilakukan untuk menebak kecepatan rambat gelombang seismik melalui medium di dalam bumi. Uniknya, gelombang seismik secara disiplin akan memberikan sinyal saat dia melalui batas-batas medium yang berbeda sifat fisikanya, sehingga bisa kita kenali masing-masing segmen medium di dalam bumi dari kenampakan jejak seismiknya. ## Terminologi Gelombang

Sebelum melangkah lebih jauh, kita perlu mengenal terminologi gelombang, misalnya
### Panjang gelombang, $\lambda$

Yaitu jarak antara 2 titik pada gelombang yang memiliki fase yang sama atau posisi perpindahan yang sama.

![](http://www.chemistry.wustl.edu/~coursedev/Online%20tutorials/waves/sinwave.gif)

Jadi jarak dari puncak ke puncak adalah 1 panjang gelombang.

### Amplitudo gelombang

Yaitu perpindahan maksimum dari gerakan partikel dalam medium saat dilalui gelombang.

Amplitudo dan panjang gelombang memberikan gambaran tentang bentuk gelombang (waveform)

### Periode

Yaitu waktu yang diperlukan untuk melakukan 1 kali siklus gelombang.

![](https://scientifichealer.com/courses/phy214/harmonicequations_files/wavesfig1b.gif)

### Frekuensi

Yaitu jumlah perulangan tiap satuan waktu. Jika gelombang seismik memiliki periode 0,10 s, maka frekuensi gelombangnya adalah 10 tiap detik atau 10 Hz. 


| Type          | Wavelength             | Frequency                |

| Infrared      | 7 × 10^–7 to 10^–3     | 3 × 10^11 to 4 × 10^14   |
| Visible Light | 4 × 10^–7 to 7 × 10^–7 | 4 × 10^14 to 7.5 × 10^14 |

| X-Rays        | 10^–11 to 10^–8        | 3 × 10^16 to 3 × 10^19   |
| Gamma Rays    | < 10^–11               | > 3 × 10^19              |
